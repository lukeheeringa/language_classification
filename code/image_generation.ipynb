```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pickle
from skimage.io import imsave

from os import listdir
```

```python
def grab_index(path):
    return path.split('/')[-1]
```

```python
def grab_lang(path):
    return path.split('/')[-2]
```

```python
en_spec = np.array(['../data/spectrogram/en/' + fname for fname in listdir('../data/spectrogram/en/') if '.pkl' in fname])
zh_spec = np.array(['../data/spectrogram/zh/' + fname for fname in listdir('../data/spectrogram/zh/') if '.pkl' in fname])
```

```python
def save_spec(path):
    with open(path, 'rb') as f:
        reloaded = pickle.load(f)
    path = path.replace('spectrogram', 'image')
    path = path.replace('.pkl', '.png')
    imsave(path, reloaded)
```
